# map() 내장 함수
iterable 객체에서 요소를 하나하나 꺼내어 특정 함수를 각각 적용하고 싶을 때<br>사용합니다. 반환은 map 클래스를 반환하는데 이는 iterable한 객체입니다.<br> list() 함수 또는 tuple() 함수를 활용하여 map 클래스를 list 혹은 tuple로 변경가능합니다.

    - map(function, iterable) -> <class 'map'>(iterable)

```python
a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int,a))
print(a)
```

```python
a = map(int,a)
a
```
