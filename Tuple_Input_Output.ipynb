## 연습 문제: 리스트와 튜플 차이 생각해보기
1. 처음 20번째까지 소수(Prime number)
-> 리스트: 수가 비교적 적은 경우여서
2. 프로그래밍 언어의 종류
-> 리스트: 수가 비교적 적은 경우 + 프로그래밍 언어가 추가되어야 할 때 수정해야해서
3. 사람의 나이, 몸무게, 키
-> 리스트: 수정이 필요한 경우
4. 사람의 생일과 출생지
-> 튜플: 수가 비교적 많은 경우 + 불변 하기에
5. 롤 게임의 결과
-> 튜플: 수가 비교적 많은 경우 + 불변 하기에
6. 잇단 롤 게임의 결과
-> 리스트: 지속적인 수정을 위해 

```python
import time

start_time = time.time() # 시작하는 시간
test = []

# 1,000,000번을 반복해서 tuple 만듭니다.
for _ in range(1_000_000):
    tp = (_,_) # tuple
    test.append(tp)

# 끝나는 시간 - 시작하는 시간
print(time.time() - start_time)
# 동작한 시간
```

```python
start_time = time.time()
test = []

for _ in range(1_000_000):
    tp = [_,_] # list
    test.append(tp)
    
print(time.time() - start_time)
```

### map() 내장 함수

## 2) 문자열의 format() 메서드

```python
print("{}".format(10))
print("{} {} {}".format(10, 20, 30))
print("{} {} {} {} {}".format(101, 202, 303, 404, 505))
```

```python
import time

today = """오늘의 날짜를 알려드립니다!
{}년 {}월 {}일입니다.\
""".format(time.localtime().tm_year,
           time.localtime().tm_mon,
           time.localtime().tm_mday)

print(today)
```

```python
# 중괄호 개수보다 format 메서드의 인수가 많으면
# 초과되는 format 메서드의 인수는 버려집니다.
some_str = "중괄호는 2개이고, {} {}".format("여기","인수는","3개 입니다.")
print(some_str)
```

### - 문자열의 f-string
-> 문자열 앞에 f 알파벳을 붙이고 중괄호 안에 리터럴 값 혹은 변수를 넣어서 표현할 수도 있습니다.
