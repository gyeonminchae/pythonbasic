```python
# for문

easy_list = [[1,2,3,4,5], [3,4]]
for item in easy_list:
    for i in item:
        print(i)
```

```python
# while 문

easy_list = [[1,2,3,4,5], [3,4]]
i = 0 # easy_list 인덱스용

while i < len(easy_list):
    j = 0 # easy_list[0], easy_list[1] 인덱스용
    while j < len(easy_list[i]):
        print(easy_list[i][j])
        j += 1
    i += 1
```

# random 라이브러리

-> 임의의 수를 생성할 때 사용하는 표준 라이브러리

random.randit(시작 정수, 끝 정수): 시작 정수(포함)부터 끝 정수(포함)까지 범위의 난수 생성

```python
import random
random_list = [random.randint(1, 2) for _ in range(15)]
print(random_list)

random_list_2 = [random.randint(1,10) for _ in range(100)]
print(random_list_2)
```

```python
# 연습 문제
random_num = random.randint(1,100)
count = 0
result_list = []

while True:
    num = int(input("1부터 100까지의 숫자 중 하나를 입력하세요.:"))
    
    if num > random_num:
        print("그보다 작습니다.")
        count += 1
        result_list.append(num)
        continue
    elif num < random_num:
        print("그보다 큽니다.")
        count += 1
        result_list.append(num)
        continue        
    elif num < 0 or num > 100:
        print("1~100사이의 숫자를 입력하세요.")
        count += 1
        result_list.append(num)
        continue        
    elif num == random_num:
        print("맞췄습니다.!")
        count += 1
        result_list.append(num)
        print(f'{count}번만에 맞추셨습니다.')
        print('입력한 숫자는 다음과 같습니다. ', end = "")
        [print(i, end = " ") for i in result_list]
        break
```

```python
# 풀이
random_num = random.randint(1,100)
input_values = []

while True:
    input_values.append(input("1~100 중 숫자를 맞춰보세요:"))
    _ = int(input_values[-1])
    if _ < 0 or _ > 100:
        print("1~100 사이의 숫자를 입력하세요.")
    elif _ < random_num:
        print("그보다 큽니다.")    
    elif _:
        print("그보다 작습니다.")
    else:
        print("맞췄습니다.")
        break
print(len(input_values), "번만에 맞추셨숩니다.", sep="")
print("입력한 숫자는 다음과 같습니다.", ", ".join(input_values))
```

## 세트 자료형 - 합집합: union() 혹은 Vertical bar(|)

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.union(A,B):", set.union(A, B))
print("A | B:", A | B)
```

## 세트 자료형 -교집합: intersection() 혹은 &

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.intersection(A,B):", set.intersection(A, B))
print("A & B:", A & B)
```

## 세트 자료형 -차집합: difference() 혹은 -

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.difference(A,B):", set.difference(A, B))
print("A - B:", A - B)
```

## 세트 자료형 -대칭차집합: symmetric_difference() 혹은 ^

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.symmetric_difference(A,B):", set.symmetric_difference(A, B))
print("A ^ B:", A ^ B)
```

## 함수의 정의(define) - 인자(parameter)

-> 함수(또는 메서드) 정의에서 함수가 받을 수 있는 변수(함수인자 혹은 매개변수)<br>
-> 함수 정의절에서 소괄호 안에 정의됩니다.<br>

* 위치-키워드(positional or keyword): 위치로 혹은 키워드로 전달될 수 있는 인자를 말합니다. 매개변수의 기본 형태입니다.<br>
* 위치 전용(positional-only): / 문자를 기준으로 좌측에 위치한 인자를 말합니다.<br>
* 키워드 전용(keyword-only): * 문자를 기준으로 우측에 위치한 인자를 말합니다.<br>
* 가변 위치(var positional): 위치 매개변수 외에 추가적인 위치 매개변수를 개수에 상관없이 받을 수 있습니다. 변수명 앞에 *를 앞에 하나 붙여 표기합니다.<br>
* 가변 키워드(var keyword): 키워드 매개변수 외에 추가적인 키워드 매개변수를 개수에 상관없이 받을 수 있습니다. 변수명 앞에 **를 두개 붙여 표기합니다.

ex) dict_a = dict(key='value', key2='value2')

### 함수의 정의 - 키워드 전용 인자(keyword-only parameters)

-> 기본값을 가질 수 있음(이때 기본값을 갖는 인자에 상응하는 인수(arguments)는 생략 가능)<br>
-> 키워드 전용 인자(keyword only parameter)는 오직 위치 인수(keyword arguments)만 값의 전달이 가능합니다. 키워드 전용 인자를 선언하는 방법은 애스터리스크(*)를 인자 값으로 넣고 우측 부분에 키워드 전용 인자를 선언

### 함수의 정의 - 가변-위치 인자(var-positional parameters)

-> <u>명시된 인자 외에</u> 추가적으로 위치 인수를 개수에 상관없이 유연하게 전달 받을수 있음. 가변 변수로 사용할 변수명 앞에 '*'를 앞에 하나 붙여 표기

```python
def var_positional(*args):
    print(type(args))
    return sum([_ for _ in args])
```

### 함수의 정의 - 가변-키워드 인자(var-keyword parameters)

-> 명시된 인자 외에 추가적으로 키워드 인수를 개수에 상관없이 유연하게 전달 받을 수 있음. 가변 변수로 사용할 변수명 앞에 '*'를 앞에 두개 붙여 표기

```python
print(var_keyword())
```

## 람다(Lambda) feat.map()

map(lambda parameter1, parameter2, ...:expression, iterable1,iterable2)

```python
a = [_ for _ in range(1,6)]
b = [_ for _ in range(2,11,2)]
print(list(map(lambda x, y: x * y, a, b)))
```

## 람다(Lambda) feat. filter()

map(lambda parameter: expression, iterable)
-> filter의 첫 인수로 들어가는 함수의 반환값이 True일 때만 해당 요소를 가져온다.

```python
a = [3, 2, 8, 22, 10, 7, 0, 11, 9, 9]
list(filter(lambda x: x % 3 == 0 and 0 < x < 10, a))
```
