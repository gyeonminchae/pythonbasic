# 리스트 컴프리헨션(for + if)

1. 반복가능한 객체(iterable) 값을 하나 꺼내와서 조건식에 넣습니다.
2. 조건식이 True이면 변수를 활용해서 표현식(expression)을 실행합니다.
    (이때 변수는 활용할 수도 활용하지 않을 수도 있습니다.)
3. 조건식이 False이면 1번부터 다시 시작
4. 표현식 결과 값을 리스트 요소에 넣습니다.
5. 반복가능한 객체(iterable)로부터 값이 없을 때까지 1~3 과정을 반복

list_variable = [expression for 변수 in iterable if 조건식]

```python
list_p = [i for i in range(10) if i % 3 == 0]
print(list_p)
```

# 중첩 리스트 사용하기

-> 리스트 안에 리스트를 포함할 수 있음. 리스트는 수학의 배열과 다름. 또 리스트를 수학의 배열처럼 사용하기에도 적절x. 따라서 리스트는 리스트 자체로 이해(중첩된 리스트를 차원이라고 여기지 x)

```python
# for문

easy_list = [[1,2,3,4,5], [3,4]]
for item in easy_list:
    for i in item:
        print(i)
```

```python
# while 문

easy_list = [[1,2,3,4,5], [3,4]]
i = 0 # easy_list 인덱스용

while i < len(easy_list):
    j = 0 # easy_list[0], easy_list[1] 인덱스용
    while j < len(easy_list[i]):
        print(easy_list[i][j])
        j += 1
    i += 1
```

# random 라이브러리

-> 임의의 수를 생성할 때 사용하는 표준 라이브러리

random.randit(시작 정수, 끝 정수): 시작 정수(포함)부터 끝 정수(포함)까지 범위의 난수 생성

```python
import random
random_list = [random.randint(1, 2) for _ in range(15)]
print(random_list)

random_list_2 = [random.randint(1,10) for _ in range(100)]
print(random_list_2)
```

```python
# 연습 문제
random_num = random.randint(1,100)
count = 0
result_list = []

while True:
    num = int(input("1부터 100까지의 숫자 중 하나를 입력하세요.:"))
    
    if num > random_num:
        print("그보다 작습니다.")
        count += 1
        result_list.append(num)
        continue
    elif num < random_num:
        print("그보다 큽니다.")
        count += 1
        result_list.append(num)
        continue        
    elif num < 0 or num > 100:
        print("1~100사이의 숫자를 입력하세요.")
        count += 1
        result_list.append(num)
        continue        
    elif num == random_num:
        print("맞췄습니다.!")
        count += 1
        result_list.append(num)
        print(f'{count}번만에 맞추셨습니다.')
        print('입력한 숫자는 다음과 같습니다. ', end = "")
        [print(i, end = " ") for i in result_list]
        break
```

```python
# 풀이
random_num = random.randint(1,100)
input_values = []

while True:
    input_values.append(input("1~100 중 숫자를 맞춰보세요:"))
    _ = int(input_values[-1])
    if _ < 0 or _ > 100:
        print("1~100 사이의 숫자를 입력하세요.")
    elif _ < random_num:
        print("그보다 큽니다.")    
    elif _:
        print("그보다 작습니다.")
    else:
        print("맞췄습니다.")
        break
print(len(input_values), "번만에 맞추셨숩니다.", sep="")
print("입력한 숫자는 다음과 같습니다.", ", ".join(input_values))
```

## 세트 자료형 - 세트에 값 넣는 방법

- add(): 세트 자료형에 값을 추가

세트 자료형.add(요소)

```python
set_variable = set()
print(set_variable, type(set_variable))
```

```python
# 정수 1 값을 set_variable에 5차례 넣는다
[set_variable.add(1) for _ in range(5)]

# 출력하면 중복된 값 없이 요소를 갖음
print(set_variable)
```

## 세트 자료형 - 합집합: union() 혹은 Vertical bar(|)

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.union(A,B):", set.union(A, B))
print("A | B:", A | B)
```

## 세트 자료형 -교집합: intersection() 혹은 &

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.intersection(A,B):", set.intersection(A, B))
print("A & B:", A & B)
```

## 세트 자료형 -차집합: difference() 혹은 -

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.difference(A,B):", set.difference(A, B))
print("A - B:", A - B)
```

## 세트 자료형 -대칭차집합: symmetric_difference() 혹은 ^

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
print("set.symmetric_difference(A,B):", set.symmetric_difference(A, B))
print("A ^ B:", A ^ B)
```

## 세트 자료형 - update() 혹은 |=

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
A.update(B) # A |= B 와 같습니다.
print("A.update(B):", A)
```

## 세트 자료형 - intersection_update() 혹은 &=

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
A.intersection_update(B) # A &= B 와 같습니다.
print("A.intersection_update(B):", A)
```

## 세트 자료형 - difference_update() 혹은 -=

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
A.difference_update(B) # A -= B 와 같습니다.
print("A.difference_update(B):", A)
```

## 세트 자료형 - symmetric_difference_update() 혹은 ^=

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}
print("A:", A, "\nB:", B)
A.symmetric_difference_update(B) # A ^= B 와 같습니다.
print("A.symmetric_difference_update(B):", A)
```
